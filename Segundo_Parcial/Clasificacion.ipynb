```python
import torch
import torchtext
from torchtext.legacy import datasets
from torchtext.legacy import data
```

```python
ID = data.LabelField(dtype = torch.long)
PRELIMINARY_DECISION = data.RawField()
REVIEW = data.RawField()
CONFIDENCE = data.LabelField(dtype = torch.long) 
EVALUATION = data.LabelField(dtype = torch.long) 
ID = data.LabelField(dtype = torch.long) 
LAN = data.RawField()
ORIENTATION = data.LabelField(dtype = torch.long)
REMARKS = data.RawField()
TEXT = data.Field(tokenize = 'spacy')
LABEL = data.RawField()
```

```python
dataset = data.TabularDataset(
    path = 'data.json',
    format = 'json',
    fields [('id', ID), ('preliminary_decision', PRELIMINARY_DECISION), ('review',REVIEW), ('confidence', CONFIDENCE), ('evaluation', EVALUATION), ('id', ID), ('lan', LAN), ('orientation', ORIENTATION), ('remarks', REMARKS), ('text', TEXT), ('timespan',LABEL)],
    skip_header =True
)
```
