```
import numpy as np
from tensorflow.python.keras import backend as k
k.clear_session()
import numpy as np
import pandas as pd
np.random.seed(10)

from keras.layers import Input, Dense, SimpleRNN
from keras.models import Model
#from keras.optimizers import SGD
#from keras.utils import to_categorical
from keras import backend as K
```

```
dataset_train = pd.read_csv("/content/gdrive/MyDrive/datasets/Temperaturas/dataTemperatura.txt")

print(dataset_train.head())

training_set = dataset_train.iloc[:, 0:1].values

print(training_set)
```

```
#normalizacion de los datos usando el metodo MinmaxScaler 
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler(feature_range = (0, 1))
training_set_scaled = sc.fit_transform(training_set)


X_train, y_train = [], []
for i in range(30, len(training_set_scaled)):
    X_train.append(training_set_scaled[i - 30: i, 0])
    y_train.append(training_set_scaled[i])

X_train, y_train = np.array(X_train), np.array(y_train)

print(f'Forma de x: {X_train.shape}')
print(f'Forma de y: {y_train.shape}')

y_train
```

```
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
print(X_train)
```

```
#entrenamiento del modelo


from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM

modelo = Sequential()
modelo.add(LSTM(40,return_sequences = True, input_shape = (X_train.shape[1], 1)))
modelo.add(Dropout(.2))
modelo.add(LSTM(40, return_sequences = True))
modelo.add(Dropout(.2))
modelo.add(LSTM(40, return_sequences = False))
modelo.add(Dense(1))
modelo.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['accuracy'])

modelo.fit(X_train, y_train, epochs = 300, batch_size = 32)
```

```
#Probar precio de Temepratura Minima
dataset_test = pd.read_csv("/content/gdrive/MyDrive/datasets/Temperaturas/dataTemperatura.txt")
real_stock_val = dataset_test.iloc[:,0:1].values
dataset_total = pd.concat((dataset_train['Tempmínima'], dataset_test['Tempmínima']), axis=0)

inputs = dataset_total[len(dataset_total) - len(dataset_test) - 30:].values
inputs = inputs.reshape(-1, 1)
inputs = sc.transform(inputs)
X_test = []
for i in range(30, len(inputs)):
    X_test.append(inputs[i - 30: i])

X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
prediccionTemperaturas = modelo.predict(X_test)
print(prediccionTemperaturas)
```
